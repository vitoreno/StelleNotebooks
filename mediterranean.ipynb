```
# Carico la libreria "pandas" per la gestione dei dati e "pyplot" per la visualizzazione dei grafici
import pandas as pd
import matplotlib.pyplot as plt

# Leggo i dati dal file "mediterranean_surface_temperature.csv"
data = pd.read_csv("/content/mediterranean_surface_temperature_2016.csv")

# Visualizzo i primi 5 record
data.head()
```

```
# Visualizzo un diagramma a torta delle percentuali di record relativi a mar Adriatico, Ionio e Tirreno  
data.sea.value_counts().plot.pie(autopct='%1.0f%%', figsize=(5,5))
plt.axis('equal')
plt.title('Records belonging to the Adriatic, Ionian and Tyrrhenian seas')
```

```
# Istogramma delle temperature medie di mar Adriatico, Ionio e Tirreno durante i mesi dell'anno 2016

# Calcolo la media della temperatura per mese e mare (Adriatico, Ionio e Tirreno)
sst_mean = data.groupby([pd.DatetimeIndex(data.time).month, data.sea]).mean()

# Visualizzo i dati
sst_mean.unstack().plot.bar(y='sst', xlabel='month', ylabel='surface temperature', ylim=[285,300], figsize=(10,5))
plt.title('Surface temperature mean over the months')
```

```
fig, axes = plt.subplots(nrows=1, ncols=2, figsize=(10,5))
data.boxplot(column='lat', by='sea', ax=axes[0])
data.boxplot(column='lon', by='sea', ax=axes[1])
```
